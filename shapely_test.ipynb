```python
import shapely
import matplotlib.pyplot as plt
from shapely.geometry import Polygon, LineString

a = LineString([(0, 0), (1, 1)])
b = LineString([(1, 1), (2, 2)])
c = Polygon([(0,0), (0,5), (5,5), (5,0)])
d = Polygon([(0,1), (0,-5), (5,-5), (5,0)])

plt.plot(*a.xy)
#plt.plot(*b.xy)
plt.plot(*c.exterior.xy)
plt.plot(*d.exterior.xy)

print(a.touches(c))
print(d.touches(c))
print(d.overlaps(c))
```
